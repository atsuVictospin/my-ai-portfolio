```python
"""
特徴量エンジニアリング（feature engineering）：生データを機械学習で扱いやすい形に変換すること。

データクレンジング（data cleansing）：欠損値やノイズを除去しモデルが読み込みやすいデータに整えること。

正規化（normalization）：画像のピクセル値を一定範囲にスケーリングする手法。

データ拡張（data augmentation）：学習データを増やすために画像を回転、反転、ズームなどで変換する技術。

平滑化、エッジ強調などのフィルタ処理（filtering）：画像の特徴を抽出しやすくするための処理。

これらの前処理は、PythonのOpenCV、PIL、torchvision.transformsやTensorFlowの前処理レイヤーを用いて実装されることが多いです。前処理を適切に行うことで画像分類AIモデルの精度向上に大きく寄与します.​

以上の英語表現を...日本語﻿の形式で強調する場合：

前処理﻿ = preprocessing or data preprocessing

特徴量エンジニアリング﻿ = feature engineering

データクレンジング﻿ = data cleansing

正規化﻿ = normalization

データ拡張﻿ = data augmentation

フィルタ﻿ = filtering

これらの用語は画像分類AIの説明で英語文書にもしばしば登場します。
"""
```

```python
#ファイル名を001からカウントアップし値に変えていく
import os  # OSのファイル操作モジュールを読み込む

train_negative_path = R"C:\work\mazda_ai_portfolio\dataset\metal_nut\metal_nut\train\negative" # 対象のフォルダパスを設定する
val_negative_path = R"C:\work\mazda_ai_portfolio\dataset\metal_nut\metal_nut\val\negative" # 対象のフォルダパスを設定する

def rename(folder_path):
    count=1 # カウンター開始
    # フォルダ内の全ファイル名を取得してループで処理
    for filename in os.listdir(folder_path):
        print(filename)
        # フォルダの中にあるそれぞれのファイル名を filename という名前で順番に取り出して使う
        base, ext = os.path.splitext(filename)  # base:ファイル名とext:拡張子に分ける
        new_name = str(count).zfill(3) + ext   # 例："001.jpg"
        #new_name = base + "5" + ext  # ファイル名の最後に"1"を追加します
        old_file = os.path.join(folder_path, filename)  # 元ファイルのフルパス作成
        new_file = os.path.join(folder_path, new_name)  # 新ファイル名のフルパス作成
        os.rename(old_file, new_file)  # ファイル名を変更
        count += 1  # カウンターを1つ増やす
        
    print(f"{folder_path}のファイルの名前変更が完了しました。")

train_negative=rename(train_negative_path)
val_negative=rename(val_negative_path)
```

```python
# Pillow＋NumPyのコードは画像ファイル自体の加工（サイズ変更・正規化後の保存）を行いたい場合に適している。

from PIL import Image  # 画像を扱うためのライブラリPillowを読み込む
import numpy as np  # 数値計算用のライブラリNumPyを読み込む
import os  # ファイル操作・フォルダ操作用のモジュールを読み込む


train_positive_path = R"C:\work\mazda_ai_portfolio\dataset\metal_nut\metal_nut\train\positive" # 画像が入っているフォルダの名前（適宜変更してください）
output_tp_path =  R"C:\work\mazda_ai_portfolio\dataset\metal_nut\metal_nut\train\positive_processed_images"  # 処理後の画像を保存するフォルダ名です

train_negative_path =  R"C:\work\mazda_ai_portfolio\dataset\metal_nut\metal_nut\train\negative" # 画像が入っているフォルダの名前（適宜変更してください）
output_tn_path =  R"C:\work\mazda_ai_portfolio\dataset\metal_nut\metal_nut\train\negative_processed_images"  # 処理後の画像を保存するフォルダ名です

val_positive_path = R"C:\work\mazda_ai_portfolio\dataset\metal_nut\metal_nut\val\positive" # 画像が入っているフォルダの名前（適宜変更してください）
output_vp_path =  R"C:\work\mazda_ai_portfolio\dataset\metal_nut\metal_nut\val\positive_processed_images"  # 処理後の画像を保存するフォルダ名です

val_negative_path =  R"C:\work\mazda_ai_portfolio\dataset\metal_nut\metal_nut\val\negative" # 画像が入っているフォルダの名前（適宜変更してください）
output_vn_path =  R"C:\work\mazda_ai_portfolio\dataset\metal_nut\metal_nut\val\negative_processed_images"  # 処理後の画像を保存するフォルダ名です


def resize(folder_path,output_folder):
    if not os.path.exists(output_folder):  # 出力用フォルダがなければ
        os.makedirs(output_folder)  # 新しくフォルダを作成します
    
    for filename in os.listdir(folder_path):  # フォルダ内のすべてのファイルに対して繰り返す
        if filename.endswith(".jpg") or filename.endswith(".png"):  # jpgまたはpngファイルだけ処理対象にする。
            # endswith() は文字列の最後が特定の文字列かどうかをチェックするメソッド
            img_path = os.path.join(folder_path, filename)  # 元画像ファイルのフルパスを作る
            img = Image.open(img_path)  # 画像を開く
            img_resized = img.resize((64, 64))  # 画像のサイズを64×64ピクセルに統一しリサイズする
            img_array = np.array(img_resized)  # 画像を数値の配列（NumPy配列）に変換する
            img_normalized = img_array / 255.0  # ピクセルの値を0〜1に正規化する（255で割る）。255は一番明るい（白）、0は一番暗い（黒）という意味
            """正規化は、「データを同じスケールに揃えて、モデルが効率よく学ぶための準備」"""
            img_normalized = (img_normalized * 255).astype(np.uint8)  # 再び255スケールに戻して整数化（保存用）
            """.astype(np.uint8) はデータ型を8ビットの符号なし整数に変換します。これは画像ファイルで一般的なピクセルの形式です。
                これをしないと小数のままで保存できなかったり、描画に問題が出ます。"""
            img_final = Image.fromarray(img_normalized)  # NumPy配列からPillowの画像オブジェクトに戻す
            save_path = os.path.join(output_folder, filename)  # 保存先のパスを作ります
            img_final.save(save_path)  # 正規化・リサイズした画像を保存する
    
    print(f"{folder_path}全画像の前処理（リサイズ・正規化）を終えました。")

train_positive_resize=resize(train_positive_path,output_tp_path)
train_negative_resize=resize(train_negative_path,output_tn_path)
val_positive_resize=resize(val_positive_path,output_vp_path)
val_negative_resize=resize(val_negative_path,output_vn_path)
```
