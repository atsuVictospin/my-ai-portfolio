```python
import torch

# 1Dテンソル（ベクトル）
x = torch.tensor([1, 2, 3])
print("x:", x)

# 2Dテンソル（行列）
y = torch.tensor([[1, 2], [3, 4]])
print("y:\n", y)

# 要素ごとの加算
print("x + 10 =", x + 10)

# 行列の掛け算
result = torch.matmul(y, y)
print("y × y =\n", result)
```

```python
import torch

# 1Dテンソル（ベクトル）
x = torch.tensor([1, 2, 3])
print("x:", x)

# 2Dテンソル（行列）
y = torch.tensor([[1, 2], [3, 4]])
print("y:\n", y)

# 要素ごとの加算
print("x + 10 =", x + 10)

# 行列の掛け算
result = torch.matmul(y, y)
print("y × y =\n", result)
```

```python
print("y の形:", y.shape)
print("y の型:", y.dtype)
print("y を NumPy 配列に変換:", y.numpy())
```

```python
# GPUを使って計算してみる
if torch.cuda.is_available():
    device = torch.device("cuda")
    x_gpu = x.to(device)
    print("GPU上のテンソル：", x_gpu)
```
